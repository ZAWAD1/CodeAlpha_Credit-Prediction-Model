_**Data Pre-processing**_

MonthlyIncome value fillups

```python
# Total number of missing values in MonthlyIncome
df['MonthlyIncome'].isnull().sum()
```

```python
# Filling missing values of MonthlyIncome with median
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
df['MonthlyIncome'].isnull().sum()
```

NumberOfDependents value fillups

```python
# Total number of missing values in NumberOfDependents
df['NumberOfDependents'].isnull().sum()
```

```python
# Filling the missing values of NumberOfDependents with median
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(df['NumberOfDependents'].median())
df['NumberOfDependents'].isnull().sum()
```

Cliping extreme outliers found in the EDA

```python
# DebtRatio cliping
df['DebtRatio'] = df['DebtRatio'].clip(upper=df['DebtRatio'].quantile(0.99))
# RevolvingUtilizationOfUnsecuredLines cliping
df['RevolvingUtilizationOfUnsecuredLines'] = df['RevolvingUtilizationOfUnsecuredLines'].clip(upper=1.5)
# MonthlyIncome cliping
df['MonthlyIncome'] = df['MonthlyIncome'].clip(upper=df['MonthlyIncome'].quantile(0.99))
```

_**Feature Engineering**_

Late payment merge by addition,
There are total 3 late payment features, but they are highly correlated. We will keep only total_late for the models.

```python
# total_late
df['total_late'] = (df['NumberOfTime30-59DaysPastDueNotWorse'] + 
                    df['NumberOfTime60-89DaysPastDueNotWorse'] + 
                    df['NumberOfTimes90DaysLate'])
df.head(2)
```

Rough debt amount estimation, by multiplying DebtRatio with MonthlyIncome for model interpretability


```python
# Rough debt amount
df['debt_amount'] = df['DebtRatio'] * df['MonthlyIncome']
df.head(2)
```

_**Spliting The Data**_

```python
from sklearn.model_selection import train_test_split
```

```python
# X and Y separation
X = df.drop('SeriousDlqin2yrs', axis=1)
Y = df['SeriousDlqin2yrs']

# train and test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
print("Train default rate:", Y_train.mean().round(3))
print("Test default rate: ", Y_test.mean().round(3))
```

_**Scaled Data**_

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

_**Model Training**_

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

Model-1: Logistic Regression

```python
lr = LogisticRegression(
    max_iter=1000,
    random_state=42,
    class_weight='balanced'
)
lr.fit(X_train_scaled, Y_train)
```

Model-2: Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced'
)
rf.fit(X_train, Y_train)
```

Model-3: Decision Tree

```python
dt = DecisionTreeClassifier(
    random_state=42,
    class_weight='balanced',
    max_depth=5
)
dt.fit(X_train, Y_train)
```

Model-4: XGboost

```python
xgb = XGBClassifier(
    n_estimators=100,
    scale_pos_weight=9,
    random_state=42,
    eval_metric='logloss'
)
xgb.fit(X_train, Y_train)
```

_**Model Testing**_

Model-1: Logistic Regression

```python
y_pred_lr = lr.predict(X_test_scaled)
y_prob_lr = lr.predict_proba(X_test_scaled)[:,1]

lcr = (classification_report(Y_test, y_pred_lr, target_names=['No Default', 'Default']))
print(lcr)
print("ROC-AUC: ", round(roc_auc_score(Y_test, y_prob_lr), 4))

ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_lr, display_labels=['No default', 'Default'])
plt.title('Confusion Matrix - LogisticRegression')
plt.ylabel("")
plt.show()
```

Model-2: Random Forest

```python
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]

rcr = (classification_report(Y_test, y_pred_rf, target_names=['No Default', 'Default']))
print(rcr)
print("ROC-AUC: ", round(roc_auc_score(Y_test, y_prob_rf), 4))

ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_rf, display_labels=['No default', 'Default'])
plt.title('Confusion Matrix - Random Forest')
plt.ylabel("")
plt.show()
```

Model-3: Dicision Tree

```python
y_pred_dt = dt.predict(X_test)
y_prob_dt = dt.predict_proba(X_test)[:, 1]

dtcr = classification_report(Y_test, y_pred_dt, target_names=['No Default', 'Default'])
print(dtcr)
print("ROC-AUC:", round(roc_auc_score(Y_test, y_prob_dt), 4))

ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_dt, display_labels=['No default', 'Default'])
plt.title('Confusion Matrix - Decision Tree')
plt.ylabel("")
plt.show()
```

Model-4: XGboost

```python
y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]

xgcr = classification_report(Y_test, y_pred_xgb, target_names=['No Default', 'Default'])
print(xgcr)
print("ROC-AUC:", round(roc_auc_score(Y_test, y_prob_xgb), 4))

ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_xgb, display_labels=['No default', 'Default'])
plt.title('Confusion Matrix - XGboost')
plt.ylabel("")
plt.show()
```

_**Best Model(XGboost)**_

XGBoost was selected as the final model after training and evaluating four algorithms — Logistic Regression, Decision Tree, Random Forest, and XGBoost — on the same test data. Unlike Random Forest which missed 85% of actual defaulters, and Logistic Regression which flagged too many safe borrowers as risky, XGBoost found the best balance between the two failure modes. Its sequential boosting approach, where each tree directly corrects the previous tree's mistakes, proved more effective than independent voting or single-tree decisions. The scale_pos_weight=9 parameter handled the severe 93/7 class imbalance more precisely than class_weight='balanced', ensuring the model treated missed defaulters as significantly more costly than false alarms. With the highest F1 score of 0.39 and ROC-AUC of 0.852 for the default class.
